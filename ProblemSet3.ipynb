```python
Q1)Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men
```

```python
import pandas as pd #1)Import the necessary libraries
url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user" #2)Import the dataset from the link provided
users=pd.read_csv(url, sep='|') #3)Assign it to a variable called users
avg_age=users.groupby('occupation')['age'].mean() #4)Discover what is the mean age per occupation

#5)Discover the Male ratio per occupation and sort it from the most to the least
Mratio=users[users['gender']=='M'].groupby('occupation').size() / users.groupby('occupation').size()
Mratio=Mratio.sort_values(ascending=False)

MinMax_age=users.groupby('occupation')['age'].agg(['min', 'max']) #6)For each occupation, calculate the minimum and maximum ages

AvgAgeSex=users.groupby(['occupation', 'gender'])['age'].mean() #7)For each combination of occupation and sex, calculate the mean age

#8)For each occupation present the percentage of women and men
gender_ocup=users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count=users.groupby(['occupation']).agg('count')
occup_gender=gender_ocup.div(occup_count, level = "occupation") * 100
print("Average Age for each Occupation:\n", avg_age)
print("\nMale Ratio for each Occupation:\n", Mratio)
print("\nMinimum and Maximum Age for each Occupation:\n", MinMax_age)
print("\nMean Age for each Occupation and Sex:\n", AvgAgeSex)
print("\nPercentage of Women and Men for each Occupation:\n", occup_gender)
```

```python
Q4)Wind Statistics
The data have been modified to contain some missing values, identified by NaN.Using pandas should make this exercise easier, in particular for the bonus question.
You should be able to perform all of these operations without using a for loop orother looping construct.
The first three columns are year, month, and day. The remaining 12 columns areaverage windspeeds in knots at 12 locations in Ireland on that day.
Step 1. Import the necessary libraries
Step 2. Import the dataset from the attached file wind.txt
Step 3. Assign it to a variable called data and replace the first 3 columns by a properdatetime index.
Step 4. Year 2061? Do we really have data from this year? Create a function to fix itand apply it.
Step 5. Set the right dates as the index. Pay attention at the data type, it should bedatetime64[ns].
Step 6. Compute how many values are missing for each location over the entirerecord.They should be ignored in all calculations below.
Step 7. Compute how many non-missing values there are in total.
Step 8. Calculate the mean windspeeds of the windspeeds over all the locations andall the times.A single number for the entire dataset.
Step 9. Create a DataFrame called loc_stats and calculate the min, max and meanwindspeeds and standard deviations of the windspeeds at each location over all thedaysA different set of numbers for each location.
Step 10. Create a DataFrame called day_stats and calculate the min, max and meanwindspeed and standard deviations of the windspeeds across all the locations at eachday.A different set of numbers for each day.
Step 11. Find the average windspeed in January for each location.Treat January 1961 and January 1962 both as January.
Step 12. Downsample the record to a yearly frequency for each location.
Step 13. Downsample the record to a monthly frequency for each location.
Step 14. Downsample the record to a weekly frequency for each location.
Step 15. Calculate the min, max and mean windspeeds and standard deviations of thewindspeeds across all locations for each week (assume that the first week starts onJanuary 2 1961) for the first 52 weeks
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import datetime

# Define the data URL
data_url = 'wind.txt'

# Read the data from the specified URL, treating consecutive whitespaces as a separator,
# and parse the first three columns (year, month, day) as dates
data = pd.read_csv(data_url, sep="\s+", parse_dates=[[0, 1, 2]])

# Function to fix the century of the year
def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

# Apply the fix_century function to the 'Yr_Mo_Dy' column
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

# Convert the 'Yr_Mo_Dy' column to datetime format
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

# Set 'Yr_Mo_Dy' as the index of the DataFrame
data = data.set_index('Yr_Mo_Dy')

# Display the number of missing values for each column
print(data.isnull().sum())

# Display the count of non-null values for each column
print(data.shape[0] - data.isnull().sum())

# Display the overall mean of the DataFrame
print(data.mean().mean())

# Create a DataFrame for location-wise statistics (min, max, mean, std)
loc_stats = pd.DataFrame()
loc_stats['min'] = data.min()  # min
loc_stats['max'] = data.max()  # max
loc_stats['mean'] = data.mean()  # mean
loc_stats['std'] = data.std()  # standard deviations

# Create a DataFrame for day-wise statistics (min, max, mean, std)
day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis=1)  # min
day_stats['max'] = data.max(axis=1)  # max
day_stats['mean'] = data.mean(axis=1)  # mean
day_stats['std'] = data.std(axis=1)  # standard deviations

# Display the mean values for the month of January
print(data.loc[data.index.month == 1].mean())

# Resample the data annually and display the mean
print(data.resample('A').mean())

# Resample the data monthly and display the mean
print(data.resample('M').mean())

# Resample the data weekly and display the mean
print(data.resample('W').mean())

# Resample the data weekly and calculate statistics for each week
weekly = data.resample('W').mean().apply(lambda x: [x.min(), x.max(), x.mean(), x.std()])

# Display statistics for weeks 2 to 52
print(weekly.loc[weekly.index[1:53]])
```

```python
Q6)Create a line plot showing the number of marriages and divorces per capita in theU.S. between 1867 and 2014. 
Label both lines and show the legend.Don't forget to label your axes!
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file into a DataFrame
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Calculate per capita rates by dividing the 'Marriages' and 'Divorces' columns by the 'Population' column
data['Marriages_per_capita'] = data['Marriages'] / data['Population']
data['Divorces_per_capita'] = data['Divorces'] / data['Population']

# Create a new figure for the plot with a specified size
plt.figure(figsize=(10, 6), facecolor='#EDCABE')

# Plot the 'Marriages_per_capita' over the years
plt.plot(data['Year'], data['Marriages_per_capita'],color='#E9B666',label='Marriages')

# Plot the 'Divorces_per_capita' over the years
plt.plot(data['Year'], data['Divorces_per_capita'],color='b',label='Divorces')

# Add labels to the axes
plt.xlabel('Year')
plt.ylabel('Per Capita Rate')

# Add a title to the plot
plt.title('Marriages and Divorces per Capita in the U.S. between 1867 and 2014')

# Add a legend to distinguish between 'Marriages' and 'Divorces'
plt.legend()

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file into a DataFrame
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Filter the data for the years 1900, 1950, and 2000
filtered_data = data[data['Year'].isin([1900, 1950, 2000])]

# Set the width and opacity for the bars
bar_width = 0.35
opacity = 0.8

# Create a figure and axis for the bar chart
fig, ax = plt.subplots(figsize=(10, 6),facecolor='#EDCABE')

# Create two sets of bars, one for marriages and one for divorces, for each selected year
bar1 = plt.bar(filtered_data['Year'] - bar_width/2, filtered_data['Marriages_per_1000'], bar_width, alpha=opacity, color='#E9B666', label='Marriages')
bar2 = plt.bar(filtered_data['Year'] + bar_width/2, filtered_data['Divorces_per_1000'], bar_width, alpha=opacity, color='b', label='Divorces')

# Add labels to the x and y axes
plt.xlabel('Year')
plt.ylabel('Count per capita')

# Add a title to the plot
plt.title('Number of Marriages and Divorces per Capita in the U.S. in 1900, 1950, and 2000')

# Set x-axis ticks to the selected years
plt.xticks(filtered_data['Year'])

# Add a legend to distinguish between 'Marriages' and 'Divorces'
plt.legend()

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```

```python
Q9)Create a pie chart showing the fraction of all Roman Emperors that wereassassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows thepercentage breakdown of the categories.
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file into a DataFrame
data = pd.read_csv('roman-emperor-reigns.csv')

# Calculate the fraction of all Roman Emperors that were assassinated
assassinated_emperors = data[data['Cause_of_Death'] == 'Assassinated']
total_emperors = len(data)
fraction_assassinated = len(assassinated_emperors) / total_emperors

# Create a figure with a specified size
plt.figure(figsize=(6, 6), facecolor='#EDCABE')

# Create a pie chart to visualize the fraction of assassinated and non-assassinated emperors
labels = ['Assassinated', 'Not Assassinated']
sizes = [fraction_assassinated, 1 - fraction_assassinated]

# Define colors for each slice
colors = ['#E9B666', '#0F4C81']

# Plot the pie chart with percentage labels and a starting angle of 140 degrees
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)

# Ensure an equal aspect ratio to make the pie chart circular
plt.axis('equal')

# Add a title to the plot
plt.title('Fraction of all Roman Emperors that were assassinated')

# Display the pie chart
plt.show()
```
